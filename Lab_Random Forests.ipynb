# Instructions


**Apply the Random Forests algorithm but this time only by upscaling the data using SMOTE.**

Note that since SMOTE works on numerical data only, we will first encode the categorical variables in this case.

```python
scaler1 = OrdinalEncoder()
```

```python
# Encoding categorical data
cat = scaler1.fit_transform(cat)
cat = pd.DataFrame(cat)
cat.columns
```

```python
# Joining the target variable with the explanatives variables to check the 
# correlation between them

cat = cat.merge(target.iloc[:, 0], left_index = True, right_index = True)
X = num.merge(cat, left_index = True, right_index = True)
```

```python
# Looking for the most correlated variables 
cor = X.corr()
cor_target = abs(cor['TARGET_B'])
relevant_features = cor_target[cor_target>0.05]
relevant_features
```

```python
X = X.loc[: , ['NGIFTALL', 'CARDGIFT', 'RFA_2F', 6]]
y = target.iloc[:, 0]
```

```python
# Scaling X using Standard Scaler

scaler2 = StandardScaler()
X = scaler2.fit_transform(X)
```

```python
# Looking the distrubution of the target variable

y.value_counts()
```

```python
#Train/Test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Correction imbalance: oversampling --> SMOTE 

sm = SMOTE(k_neighbors=3)
X_train_SMOTE, y_train_SMOTE = sm.fit_resample(X_train, y_train)
```

```python
y_train_SMOTE.value_counts()
```

```python
print(X_train.shape)
print(X_train_SMOTE.shape)
```

```python
model1 = RandomForestClassifier()

def Building_model_SMOTE(model):

    model.fit(X_train_SMOTE, y_train_SMOTE)
        
    # Evaluation of the model
    pred_train_SMOTE = model.predict(X_train_SMOTE)
    pred_test_SMOTE = model.predict(X_test)
    
    # Accuracy of the model
    print('The accuracy of the model in the train set (SMOTE) is: %.3f' % accuracy_score(y_train_SMOTE, pred_train_SMOTE))
    print('The accuracy of the model in the test set (SMOTE) is: %.3f' % accuracy_score(y_test, pred_test_SMOTE))
```

```python
Building_model_SMOTE(model1)
```
